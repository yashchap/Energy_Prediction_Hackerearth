```python
#plt.plot(train['Energy'][1:50])
#plt.plot(train['T1'][1:50]*train['RH_1'][1:50])
dftemp = pd.DataFrame([])
dftemp['Energy'] = train['Energy']
dftemp['Press_mm_hg'] = train['Press_mm_hg']
dftemp['X_out'] = train['T_out']/train['RH_out']
dftemp['Windspeed'] = train['Windspeed']
#dftemp['Tdewpoint'] = train['Tdewpoint']
for i in range(1,10):
    dftemp['X'+str(i)] = train['T'+str(i)]/train['RH_'+str(i)]
train = dftemp
#sns.heatmap(dftemp.corr())
```

```python
estimators = np.arange(210, 270, 10)
scores = []
for n in estimators:
    model.set_params(n_estimators=n)
    model.fit(extra_train_x, extra_train_y)
    temp = model.score(train_x, train_y)
    scores.append(temp)
    print("Built tree with n: ",n," = ",temp )
plt.title("Effect of n_estimators")
plt.xlabel("n_estimator")
plt.ylabel("score")
plt.plot(estimators, scores)
```
