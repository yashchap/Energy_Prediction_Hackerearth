```python
train = train.append(extra_train)
train.shape
```

```python
train_y = train[['Energy']]
train_x = train.drop(['Energy'],axis=1)
```

```python
#plt.plot(train['Energy'][1:50])
#plt.plot(train['T1'][1:50]*train['RH_1'][1:50])
dftemp = pd.DataFrame([])
dftemp['Energy'] = train['Energy']
dftemp['Press_mm_hg'] = train['Press_mm_hg']
dftemp['X_out'] = train['T_out']/train['RH_out']
dftemp['Windspeed'] = train['Windspeed']
#dftemp['Tdewpoint'] = train['Tdewpoint']
for i in range(1,10):
    dftemp['X'+str(i)] = train['T'+str(i)]/train['RH_'+str(i)]
train = dftemp
#sns.heatmap(dftemp.corr())
```

```python
#plt.plot(train['Energy'][1:50])
#plt.plot(train['T1'][1:50]*train['RH_1'][1:50])
dftemp = pd.DataFrame([])

dftemp['Press_mm_hg'] = test['Press_mm_hg']
dftemp['X_out'] = test['T_out']/test['RH_out']
dftemp['Windspeed'] = test['Windspeed']
#dftemp['Tdewpoint'] = test['Tdewpoint']
for i in range(1,10):
    dftemp['X'+str(i)] = test['T'+str(i)]/test['RH_'+str(i)]
test = dftemp
#sns.heatmap(dftemp.corr())
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.25, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
sns.heatmap(train.corr())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaler = scaler.transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
model = RandomForestRegressor(n_jobs=-1)
```

```python
estimators = np.arange(30, 210, 10)
scores = []
for n in estimators:
    model.set_params(n_estimators=n)
    model.fit(X_train, y_train)
    temp = model.score(X_test, y_test)
    scores.append(temp)
    print("Built tree with n: ",n," = ",temp )
plt.title("Effect of n_estimators")
plt.xlabel("n_estimator")
plt.ylabel("score")
plt.plot(estimators, scores)
```

```python
model.set_params(n_estimators=130,max_depth=35)
model.fit(train_x, train_y)
model.score(train_x,train_y)
```

```python
pred = model.predict(test)
pred.shape
```
